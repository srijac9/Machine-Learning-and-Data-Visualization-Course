#### Activity 1:  Plot a Graph with $y = \{\log_{10} x, \log_{10} 2x, \log_{10} 3x, \log_{10} 5x, \log_{10} 10x \}$


1. Take list `x` and store number 1 to 99.

2. Calculate corresponding values for $y = \{\log_{10} x, \log_{10} 2x, \log_{10} 3x, \log_{10} 5x, \log_{10} 10x \}$
and store them in variables `y1`, `y2`,`y3`, `y4`, and `y5` respectively.

3. Plot the graph and make your observations.

```
# Import required libraries.
import numpy as np
import pandas as pd


# Take input for x-axis.
x = pd.Series([i for i in range(1,99)])

# Evaluate log(x) with base 10 for corresponding values of x.
y1 = np.log(x)/np.log(10)

# Evaluate log(2x) with base 10 for corresponding values of x.
y2 = np.log(x*2)/np.log(10)

# Evaluate log(3x) with base 10 for corresponding values of x.
y3 = np.log(3*x)/np.log(10)

# Evaluate log(5x) with base 10 for corresponding values of x.
y4 = np.log(5*x)/np.log(10)

# Evaluate log(10x) with base 10 for corresponding values of x.
y5 = np.log(10*x)/np.log(10)
```

```
# Plot the graphs
import matplotlib.pyplot as plt

plt.figure(figsize = (12, 6), dpi = 96)
# Label the x-axis
plt.xlabel('Numbers 1 to 99', fontsize = 12)
# Label the y-axis
plt.ylabel('log x', fontsize = 12)
# Add a title to the graph
plt.title("Graph with  y={log10x,log102x,log103x,log105x,log1010x}")
# Show a legend on the plot

# Function to show the plot
plt.scatter(x, y1, label='log10(x)')
plt.scatter(x, y2, label='log10(2x)')
plt.scatter(x, y3, label='log10(3x)')
plt.scatter(x, y4, label='log10(5x)')
plt.scatter(x, y5, label='log10(10x)')
plt.legend()
plt.show()
```

**Q:** What are your observations?

**A:** They all have the same shape, it just seems as if the graph has had a vertical shift upwards

#### Activity 2: Plot a graph with $y=\{\log_{1.1} x,\log_2 x,\log_3 x,\log_5 x,\log_{10} x\}$

Calculate corresponding values for $y=\{\log_{1.1} x,\log_2 x,\log_3 x,\log_5 x,\log_{10} x\}$
and store them in variables `y1`, `y2`,`y3`, `y4`, and `y5` respectively.



```
# Evaluate log(x) with base 1.1 for corresponding values of x.

y1 = np.log(x) / np.log(1.1)
# Evaluate log(x) with base 2 for corresponding values of x.
y2 = np.log(x) / np.log(2)
# Evaluate log(x) with base 3 for corresponding values of x.
y3 = np.log(x) / np.log(3)

# Evaluate log(x) with base 5 for corresponding values of x.
y4 = np.log(x) / np.log(5)

# Evaluate log(x) with base 10 for corresponding values of x.
y5 = np.log(x) / np.log(10)
```

```
# Plot the graphs
plt.figure(figsize = (12, 6), dpi = 96)
# Label the x-axis
plt.xlabel('Numbers 1 to 99', fontsize = 12)
# Label the y-axis
plt.ylabel('log x with different bases', fontsize = 12)
# Add a title to the graph
plt.title("graph with  y={log1.1x,log2x,log3x,log5x,log10x}")
# Show a legend on the plot

# Function to show the plot
plt.scatter(x, y1, label='log1.1(x)')
plt.scatter(x, y2, label='log2(x)')
plt.scatter(x, y3, label='log3(x)')
plt.scatter(x, y4, label='log5(x)')
plt.scatter(x, y5, label='log10(x)')
plt.legend()
plt.show()
```

**Q:** What are your observations?

**A:** They look similar to the first graph except for log base 1.1 (x)

**Q:** Why do you think we used $1.1$ in the question rather than $1$ for comparison?

**A:** Because log base 1 of x is undefined

#### Activity 3: Plot a graph with $y=\{\log_e x,x,x^2\}$

Calculate corresponding values for $y=\{\log_e x,x,x^2\}$
and store them in variables `y1`, `y2`, and `y3`respectively.

```
# Evaluate log(x) for corresponding values of x.

y1 = np.log(x)
# Evaluate x for corresponding values of x.
y2 = x

# Evaluate x raised to the power x for corresponding values of x.

y3 = x**2
```

```
# Plot the graphs
plt.figure(figsize = (12, 6), dpi = 96)
# Label the x-axis
plt.xlabel('Numbers 1 to 99', fontsize = 12)
# Label the y-axis
plt.ylabel('Value', fontsize = 12)
# Add a title to the graph
plt.title("graph with  y={logex,x,x2}")
# Show a legend on the plot

# Function to show the plot
plt.scatter(x, y1, label='log e (x)')
plt.scatter(x, y2, label='x')
plt.scatter(x, y3, label='x^2')
plt.legend()
plt.show()
```

**Q:** What are your observations?

**A:** x and the natural log of x are similar and linear, where x^2 is a quadratic

#### Activity 4: Plot a graph with $y=\{\log_e x, \log_e (x+4), \log_e x+4\}$

Calculate corresponding values for $y=\{\log_e x, \log_e (x+4), \log_e x+4\}$
and store them in variables `y1`, `y2`, and `y3`respectively.


```
# Evaluate log(x) for corresponding values of x.
y1 = np.log(x)
# Evaluate log(x + 4) for corresponding values of x.
y2 = np.log(x +4)
# Evaluate log(x) + 4 for corresponding values of x.
y3 = np.log(x) + 4
```

```
# Plot the graphs
plt.figure(figsize = (12, 6), dpi = 96)
# Label the x-axis
plt.xlabel('Numbers 1 to 99', fontsize = 12)
# Label the y-axis
plt.ylabel('Value', fontsize = 12)
# Add a title to the graph
plt.title(" graph with  y={logex,loge(x+4),logex+4}")
# Show a legend on the plot

# Function to show the plot
plt.scatter(x, y1, label='log e (x)')
plt.scatter(x, y2, label='log(x+4)')
plt.scatter(x, y3, label='log(x) +4')
plt.legend()
plt.show()
```

**Q:** What are your observations?

**A:** log e x and log x + 4 are same because the log x +4 is just translated upwards by 4, the log (x+4) and log x + 4 graphs are not exactly similar but seem to follow a similar ending

#### Activity 5: Plot a graph with $y=\{\log_e x,(\log_e (x))^2,\log_e (x^2)\}$

Calculate corresponding values for $y=\{\log_e x,(\log_e (x))^2,\log_e (x^2)\}$
and store them in variables `y1`, `y2`, and `y3`respectively.

```
# Evaluate log(x) for corresponding values of x.
y1 = np.log(x)

# Evaluate log(x) raised to the power 2 for corresponding values of x.
y2 = np.log(x)**2

# Evaluate log of x-squared for corresponding values of x.
y3 = np.log(x**2)
```

```
# Plot the graphs
plt.figure(figsize = (12, 6), dpi = 96)
# Label the x-axis
plt.xlabel('Numbers 1 to 99', fontsize = 12)
# Label the y-axis
plt.ylabel('Value', fontsize = 12)
# Add a title to the graph
plt.title("graph with  y={logex,(loge(x))2,loge(x2)}")
# Show a legend on the plot

# Function to show the plot
plt.scatter(x, y1, label='log e (x)')
plt.scatter(x, y2, label='log(x)^2')
plt.scatter(x, y3, label='log(x^2)')
plt.legend()
plt.show()
```

**Q:** What are your observations?

**A:** log(x^2) and log x ^2 are not similar, log x ^2 is significantly larger, while log(x^2) and log(x) are similar in shape
